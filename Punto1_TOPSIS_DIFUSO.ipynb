```python
crit = ['Eficiencia', 'prácticidad', 'fácilidad de uso', 
        'ahorro de tiempo', 'comunicación', 'Confianza', 
        'Seguridad', 'Familiar', 'Experiencia', 'Competencia']
alt = ['Colombia Yappo', 'PCHogar', 'MedMall', 'eBay', '']
alt.append('pesos')
```

Escala de valores lingüistícos=


inaceptable     :   [1.1.2]

muy malo        :   [1,2,3]       

malo            :   [2,3,4]

medio           :   [4,5,6]

bueno           :   [6,7,8]

muy bueno       :   [7,8,9]

```python
ColYa = ['bueno', 'malo', 'excelente', 'muy malo', 'muy bueno', 'muy bueno', 'medio', 'inaceptable', 'bueno', 'excelente']
PCH = ['medio', 'malo', 'excelente', 'excelente', 'inaceptable', 'muy bueno', 'medio', 'bueno', 'muy malo', 'muy bueno']
MMall = ['excelente', 'muy malo', 'muy malo', 'muy malo', 'muy bueno', 'medio', 'inaceptable', 'excelente', 'medio', 'bueno']
ebay = ['muy bueno', 'bueno', 'muy bueno', 'excelente', 'inaceptable', 'medio', 'excelente', 'bueno', 'malo', 'muy bueno']
sep = ['-','-','-','-','-','-','-','-','-','-',]
Pesos = ['excelente', 'bueno', 'malo', 'excelente', 'malo', 'muy bueno', 'bueno', 'inaceptable', 'muy malo', 'muy bueno']
```

```python
MTfuzzy = []
MTfuzzy.append(ColYa)
MTfuzzy.append(PCH)
MTfuzzy.append(MMall)
MTfuzzy.append(ebay)
MTfuzzy.append(sep)
MTfuzzy.append(Pesos)

MTfuzzyFrame = pand.DataFrame(MTfuzzy, index= alt, columns= crit)  
display(MTfuzzyFrame)
```

```python
I = sp.Matrix([1,1,2])
MM = sp.Matrix([1,2,3])       
M = sp.Matrix([2,3,4])
H = sp.Matrix([4,5,6])
B = sp.Matrix([6,7,8])
MB = sp.Matrix([7,8,9])
E = sp.Matrix([8,9,9])

NTfuzzy = MTfuzzy
for i in range(0,6,1):
    for j in range(0,10,1):
        if MTfuzzy[i][j] == 'inaceptable':
            NTfuzzy[i][j] = I.T
        elif MTfuzzy[i][j] == 'muy malo':
            NTfuzzy[i][j] = MM.T
        elif MTfuzzy[i][j] == 'malo':
            NTfuzzy[i][j] = M.T
        elif MTfuzzy[i][j] == 'medio':
            NTfuzzy[i][j] = H.T
        elif MTfuzzy[i][j] == 'bueno':
            NTfuzzy[i][j] = B.T
        elif MTfuzzy[i][j] == 'muy bueno':
            NTfuzzy[i][j] = MB.T
        elif MTfuzzy[i][j] == 'excelente':
            NTfuzzy[i][j] = E.T
NTfuzzyFrame = pand.DataFrame(NTfuzzy, index= alt, columns= crit)  
display(NTfuzzyFrame)
```

```python
c = []
NormFuzzy = NTfuzzy
for i in range(0,10,1):
    c = []
    for j in range(0,4,1):
        c.append(NTfuzzy[j][i][0,2])
    c = max(c) 
    for k in range(0,4,1):
        NormFuzzy[k][i] = NTfuzzy[k][i]/c
NormFuzzyFrame = pand.DataFrame(NormFuzzy, index= alt, columns= crit)  
display(NormFuzzyFrame)
```

```python
PondFuzzy = NormFuzzy
for i in range(0,10,1):
    for j in range(0,4,1):
        for k in range(0,3,1):
            PondFuzzy[j][i][0,k] = NormFuzzy[j][i][0,k] * NormFuzzy[5][i][0,k]

PondFuzzyFrame = pand.DataFrame(PondFuzzy, index= alt, columns= crit)  
display(PondFuzzyFrame)
```

```python
A = [[],[]]
for i in range(0,10,1):
    c = []
    ind = 0
    for j in range(0,4,1):
        c.append(PondFuzzy[j][i][0,2])
    c = max(c)
    cont = []
    for j in range(0,4,1):
        if c == PondFuzzy[j][i][0,2]:
            cont.append(j)
    if len(cont) > 1:
        c = []
        for k in cont:
            c.append(PondFuzzy[k][i][0,1])
        c = max(c)
        cont = []
        for j in range(0,4,1):
            if c == PondFuzzy[j][i][0,1]:
                cont.append(j)
        if len(cont) > 1:
            c = []
            for k in cont:
                c.append(PondFuzzy[k][i][0,0])
            c = max(c)
            for k in cont:
                if PondFuzzy[k][i][0,0] == c:
                    ind = k
        elif len(cont) == 1:
            ind = cont[0]
    elif len(cont) == 1:
        ind = cont[0]
    A[0].append(PondFuzzy[ind][i])
for i in range(0,10,1):
    c = []
    ind = 0
    for j in range(0,4,1):
        c.append(PondFuzzy[j][i][0,2])
    c = min(c)
    cont = []
    for j in range(0,4,1):
        if c == PondFuzzy[j][i][0,2]:
            cont.append(j)
    if len(cont) > 1:
        c = []
        for k in cont:
            c.append(PondFuzzy[k][i][0,1])
        c = min(c)
        cont = []
        for j in range(0,4,1):
            if c == PondFuzzy[j][i][0,1]:
                cont.append(j)
        if len(cont) > 1:
            c = []
            for k in cont:
                c.append(PondFuzzy[k][i][0,0])
            c = min(c)
            for k in cont:
                if PondFuzzy[k][i][0,0] == c:
                    ind = k
        elif len(cont) == 1:
            ind = cont[0]
    elif len(cont) == 1:
        ind = cont[0]
    A[1].append(PondFuzzy[ind][i])
id = ['ideal', 'no ideal']
AFrame = pand.DataFrame(A, index = id, columns = crit)
display(AFrame)
```

```python
Dideal = sp.zeros(4,10)
Dnoideal = sp.zeros(4,10)
for i in range(0,10,1):
    for j in range(0,4,1):
        di = 0
        dn = 0
        for k in range(0,3):
            di = di + (PondFuzzy[j][i][0,k] - A[0][i][0,k])**2
            dn = dn + (PondFuzzy[j][i][0,k] - A[1][i][0,k])**2
        Dideal[j,i] = di**(1/2)
        Dnoideal[j,i] = dn**(1/2)
alt.pop(5)
alt.pop(4)
DidealList = Dideal.tolist()
DnoidealList = Dnoideal.tolist()
DidealFrame = pand.DataFrame(DidealList, index = alt, columns = crit)
DnoidealFrame = pand.DataFrame(DnoidealList, index = alt, columns = crit)
display(DidealFrame)
display(DnoidealFrame)
```

```python
Dalt = sp.zeros(4,2)
for i in range(0,4,1):
    dideal = 0
    dnoideal = 0
    for j in range(0,10,1):
        dideal = dideal + Dideal[i,j] 
        dnoideal = dnoideal + Dnoideal[i,j]
    Dalt[i,0] = dideal
    Dalt[i,1] = dnoideal
DaltList = Dalt.tolist()
DaltListFrame = pand.DataFrame(DaltList, index = alt, columns = id)
display(DaltListFrame)
```

```python
CC = sp.zeros(4,1)
for i in range(0,4,1):
    CC[i,0] = Dalt[i,1]/(Dalt[i,1]+Dalt[i,0])

CCList = CC.tolist()
CCListFrame = pand.DataFrame(CCList, index = alt, columns = ['Coeficiente de Cercanía'])
CCListFrame = CCListFrame.sort_values(by = 'Coeficiente de Cercanía', ascending = False)
display(CCListFrame)
```
