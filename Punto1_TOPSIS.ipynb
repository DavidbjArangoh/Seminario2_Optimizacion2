```python
crit = ['Eficiencia', 'prácticidad', 'fácilidad de uso', 
        'ahorro de tiempo', 'comunicación', 'Confianza', 
        'Seguridad', 'Familiar', 'Experiencia', 'Competencia']

alt = ['Colombia Yappo', 'PCHogar', 'MedMall', 'eBay']
alt.append('pesos')

alt1 = []
for i in range(0,4,1):
        alt1.append('A'+str(i+1))
alt1.append('W')
crit1 = []
for i in range(0,10,1):
        crit1.append('C'+str(i+1))
```

```python
ColYa = [7, 3, 9, 2, 8, 8, 4, 1, 7, 9]
PCH = [4, 3, 9, 9, 1, 8, 5, 7, 2, 8]
MMall = [9, 2, 2, 2, 8, 5, 1, 9, 6, 7]
ebay = [8, 7, 8, 9, 1, 6, 9, 7, 3, 8]
pesos = [0.3, 0.05, 0.05, 0.05, 0.025, 0.1, 0.3, 0.02, 0.075, 0.03]

ColYaFrame = pand.DataFrame(ColYa, index = crit, columns = ['Colombia Yappo'])
PCHFrame = pand.DataFrame(PCH, index = crit, columns = ['PCHogar'])
MMallFrame = pand.DataFrame(MMall, index = crit, columns = ['Medmall'])
ebayFrame = pand.DataFrame(ebay, index = crit, columns = ['ebay'])
pesosFrame = pand.DataFrame(pesos, index = crit, columns = ['pesos'])

display(ColYaFrame)
display(PCHFrame)
display(MMallFrame)
display(ebayFrame)
display(pesosFrame)

#print(ColYaFrame.to_latex())
#print(PCHFrame.to_latex())
#print(MMallFrame.to_latex())
#print(ebayFrame.to_latex())
#print(pesosFrame.to_latex())
```

```python
MTopsis = sym.zeros(5,10)
for i in range(0,10,1):
    MTopsis[0,i] = ColYa[i]
    MTopsis[1,i] = PCH[i] 
    MTopsis[2,i] = MMall[i]
    MTopsis[3,i] = ebay[i]
    MTopsis[4,i] = pesos[i]

ATopsis = MTopsis.tolist()
MTopsisFrame = pand.DataFrame(ATopsis, index = alt1 ,columns = crit1)
display(MTopsisFrame)
#print(MTopsisFrame.to_latex())
```

```python
VTopsis = MTopsis
norm = 0
for j in range(0,10,1):
    norm = 0
    for i in range(0,4,1):
        norm = norm + MTopsis[i,j]**2
    norm = norm**(1/2)
    for k in range(0,4,1):
        VTopsis[k,j] = MTopsis[k,j]/norm

ATopsis = VTopsis.tolist()
VTopsisFrame = pand.DataFrame(ATopsis, index = alt1 ,columns = crit1)
display(VTopsisFrame)
#print(VTopsisFrame.to_latex())
```

```python
WTopsis = VTopsis
for i in range(0,10,1):
    for j in range(0,4,1):
        WTopsis[j,i] = VTopsis[j,i]*VTopsis[4,i]

ATopsis = WTopsis.tolist()
WTopsisFrame = pand.DataFrame(ATopsis, index = alt1 ,columns = crit1)
display(WTopsisFrame)
#print(WTopsisFrame.to_latex())
```

```python
A = sym.zeros(2,10)
for i in range(0,10,1):
    A[0,i] = max(WTopsis[0:4,i].tolist())
    A[1,i] = min(WTopsis[0:4,i].tolist())
ind = ['ideal', 'no-ideal']

ATopsis = A.tolist()
idealFrame = pand.DataFrame(ATopsis, index = ind ,columns = crit1)
display(idealFrame)
#print(idealFrame.to_latex())
```

```python
distancias = sym.zeros(4,2)
d_plus = 0
d_minus = 0
for i in range(0,4,1):
    d_plus = 0
    d_minus = 0
    for j in range(0,10,1):
        d_plus = d_plus + (A[0,j]-WTopsis[i,j])**2
        d_minus = d_minus + (A[1,j]-WTopsis[i,j])**2
    distancias[i,0] = d_plus**(1/2)
    distancias[i,1] = d_minus**(1/2)

alt1.pop(4)
ATopsis = distancias.tolist()
distancesFrame = pand.DataFrame(ATopsis, index = alt1, columns = ind)
display(distancesFrame)
#print(distancesFrame.to_latex())
```

```python
RS = sym.zeros(4,1)
for i in range(0,4,1):
    RS[i,0] = distancias[i,1]/(distancias[i,0]+distancias[i,1])
alt.pop(4)
ATopsis = RS.tolist()
RSFrame = pand.DataFrame(ATopsis, index = alt, columns = ['Coeficiente de cercanía'])
RSFrame = RSFrame.sort_values(by = 'Coeficiente de cercanía', ascending = False)
display(RSFrame)
#print(RSFrame.to_latex())
```
