```python
import random as rand
import pandas as pand 
import sympy as sym
```

ahorro de tiempo        c4 = 0,05

Confianza               c6 = 0,1

Experiencia             c9 = 0,075  
 
Competencia             c10 = 0,03

```python
crit = ['Eficiencia', 'prácticidad', 'fácilidad de uso', 
        'ahorro de tiempo', 'comunicación', 'Confianza', 
        'Seguridad', 'Familiar', 'Experiencia', 'Competencia']

alt = ['Colombia Yappo', 'PCHogar', 'MedMall', 'eBay']
alt.append('pesos')
```

```python
MTopsis = sym.zeros(5,10)
for i in range(0,10,1):
    for j in range(0,4,1):
        MTopsis[j,i] = rand.randrange(0,100)

pesos = [0.3, 0.05, 0.05, 0.05, 0.025, 0.1, 0.3, 0.02, 0.075, 0.03]
for i in range(0,10,1):
    MTopsis[4,i] = pesos[i]

ATopsis = MTopsis.tolist()
MTopsisFrame = pand.DataFrame(ATopsis, index = alt ,columns = crit)
display(MTopsisFrame)
```

```python
VTopsis = MTopsis
norm = 0
for j in range(0,10,1):
    norm = 0
    for i in range(0,4,1):
        norm = norm + MTopsis[i,j]**2
    norm = norm**(1/2)
    for k in range(0,4,1):
        VTopsis[k,j] = MTopsis[k,j]/norm

ATopsis = VTopsis.tolist()
VTopsisFrame = pand.DataFrame(ATopsis, index = alt ,columns = crit)
display(VTopsisFrame)
```

```python
WTopsis = VTopsis
for i in range(0,10,1):
    for j in range(0,4,1):
        WTopsis[j,i] = VTopsis[j,i]*VTopsis[4,i]

ATopsis = WTopsis.tolist()
WTopsisFrame = pand.DataFrame(ATopsis, index = alt ,columns = crit)
display(WTopsisFrame)
```

```python
A = sym.zeros(2,10)
for i in range(0,10,1):
    A[0,i] = max(WTopsis[0:4,i].tolist())
    A[1,i] = min(WTopsis[0:4,i].tolist())
ind = ['ideal', 'no-ideal']

ATopsis = A.tolist()
idealFrame = pand.DataFrame(ATopsis, index = ind ,columns = crit)
display(idealFrame)
```

```python
distancias = sym.zeros(4,2)
d_plus = 0
d_minus = 0
for i in range(0,4,1):
    d_plus = 0
    d_minus = 0
    for j in range(0,10,1):
        d_plus = d_plus + (A[0,j]-WTopsis[i,j])**2
        d_minus = d_minus + (A[1,j]-WTopsis[i,j])**2
    distancias[i,0] = d_plus**(1/2)
    distancias[i,1] = d_minus**(1/2)

alt.pop(4)
ATopsis = distancias.tolist()
distancesFrame = pand.DataFrame(ATopsis, index = alt, columns = ind)
display(distancesFrame)
```

```python
RS = sym.zeros(4,1)
for i in range(0,4,1):
    RS[i,0] = distancias[i,1]/(distancias[i,0]+distancias[i,1])

ATopsis = RS.tolist()
RSFrame = pand.DataFrame(ATopsis, index = alt, columns = ['Coeficiente de cercanía'])
RSFrame = RSFrame.sort_values(by = 'Coeficiente de cercanía', ascending = False)
display(RSFrame)
```
